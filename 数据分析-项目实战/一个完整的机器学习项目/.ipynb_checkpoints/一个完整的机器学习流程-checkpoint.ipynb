```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import Imputer
import seaborn as sns

from scipy.stats import norm,skew
from sklearn.model_selection import train_test_split

%matplotlib	inline
```

```python
housing = pd.read_csv('datasets/housing/housing.csv')
housing.head()
```

```python
housing.info()
```

###  初窥数据 - 绘制柱状图
    
   可以看出数据的分布偏长尾分布

```python
housing.hist(bins=50,figsize=(25,16))
plt.show()
```

#### 地理数据可视化

```python
housing.plot(kind="Scatter",x='longitude',y='latitude')
```

#### 通过加透明度可以看出分布的密度关系

```python
housing.plot(kind="Scatter",x='longitude',y='latitude',alpha=0.1)
```

#### 分析影响房价的多个因素

    参数解释：
    
        s为标记大小，也就是散点图的散点大小
        
        c为需要标记颜色的值，这里指定 median_house_value 为右侧的渐变线

```python
housing.plot(kind='Scatter',x='longitude',y='latitude',alpha=0.4,s=housing['population']/100,label='population',c='median_house_value',
             cmap=plt.get_cmap('jet'),colorbar=True,figsize=(20,14))
```
